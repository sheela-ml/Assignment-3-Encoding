# Perform Encoding for Categorical Columns(Sex, Smoker, Region)

```python
#replace()
df.head()
```

```python
sex1 = df.sex.unique()
sex1
```

```python
df.sex = df.sex.replace(['female', 'male'], [1,2])
```

```python
df.head()
```

```python
smoker1 = df.smoker.unique()
smoker1
```

```python
df.smoker = df.smoker.replace(['yes', 'no'], [1,2])
```

```python
df.head()
```

```python
region1 = df.region.unique()
region1
```

```python
df.region = df.region.replace(['southwest', 'southeast', 'northwest', 'northeast'], [1,2,3,4])
```

# Label Encoding Using Loop

```python
from pandas.core.dtypes.common import is_numeric_dtype
```
